```python
# which of these are palindromes?
words = ['bib', 'nun', 'civic', 'deified', 'rotator','stats','tenet']
rev_words = []
for word in words:
    rev = ""
    for i in word:
        rev = i + rev
    rev_words.append(rev)

# # Check which words are palindromes
# for i, word in enumerate(words):
#     if word == rev_words[i]:
#         print(f"{word} is a palindrome")

for word in words:
    for j in rev_words:
        if word == j:
            print(f"{word} is a palindrome")
```

```python
# which of these are palindromes?
words = ['bib', 'nun', 'civic', 'deified', 'rotator','stats','tenet','apple','car']
for word in words:
    rev = ""
    for i in word:
        rev = i + rev
    if word == rev:
        print(word ,"is pallindrome" )
```

```python
# Exercise:
# Following are synonyms of awesome as given by thesaurus.com
# Find the shortest and longest synonyms
```

```python
synonyms = ["amazing","astonishing","astounding","awe-inspiring","breathtaking","imposing","inspiring","magnificent","majestic","miraculous","spectacular","staggering","stunning","stupefying","sublime","wonderful","wondrous","affecting","arresting","august","exalted","formidable","grand","moving","overwhelming","shocking","solemn","striking","daunting","dreadful","fearsome","intimidating","transcendent"]

shortest = synonyms[0]
longest = synonyms[0]
for word in synonyms:
    if len(word)<len(shortest):
        shortest = word
    if len(word)>len(longest):
        longest = word

print(f"Shortest synonym is {shortest} and longest synonym is {longest}")   
```

```python
# vowels are a,e,i,o,u
# check how many vowels are present in a given text
```

```python
text = "hello there"
count = 0
for i in text:
    if (i == 'a' or i == 'e' or i == 'i' or i == 'o' or i == 'u' ):
        count += 1
print("there are", count, "vowels in this text")
```

```python
text = "hello there"
count = 0
space_count =0
for i in text:
    if (i == 'a' or i == 'e' or i == 'i' or i == 'o' or i == 'u' ):
        count += 1
    if i==" ":
        space_count+=1

print(f"there are {count} vowels,{space_count} space(s) and {len(text)-count-space_count} constants in this text")
```

```python
password="hyRu0876w"
entered_password=input("Enter your password: ")
#method 2
if len(password) == len(entered_password):
    match = True
    for i in range(len(password)):
        if password[i] != entered_password[i]:
            match = False
            break
    if match:
        print("Password is correct")
    else:
        print("Password is incorrect")
else:
    print("Password is incorrect")
```
