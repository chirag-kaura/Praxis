```python
# function can recursively call itself

def recurs(j):
  if(j>0):
    print(j)
    recurs(j-1)
  else:
    print("end of recurs")
  return 
```

# *args
this allows function to receive variable number of arguments, these are just values passed

some functions may have behavior which can be modified by supplying additional, optional, parameters. In such cases *args
helps to handle variable # of parameters.

```python
def sum_of(x,y,z):
    print(x+y+z)
    return x+y+z
# this function expects exactly three parameters.
# what if we want flexibility?
```

```python
sum_of(1,2,3,4)
```

```python
def total_these(*nums):
    print(nums)
    return(sum(nums))
```

```python
    
total_these(2,3,4,5,6)
```

```python
def greet (*names):
   print(names)
   for i in names:
       print("Hello", i)
    #    break
```

```python
greet("Raj", "Kumar")
```

```python
def sum_of(*nums):
    sum = 0
    print(type(nums))
    for n in nums:
        sum = sum + n

    print("sum of input numbers is:",sum)
```

```python
# sum_of(3,5)
# sum_of(4,5,6,7)
sum_of(1,2,3,5,6)
```

```python
# similarly, parameters can be of mixed types also
def intro(*data):
    for d in data:
        print(d)
```

```python
intro("Sita","Sharma",22, 1234567890)
```

```python
def intro(*data):
    print(data[0])
    print(data[1])
```

# **kwargs
this allows function to receive variable number of arguments which are assigned keywords like a variable name

```python
# this expects keywords i.e. variable names to be associated with each input value. It constructs a dictionary object.
```

```python
def intro(**data):
    print("\nData type of argument:",type(data))
```

```python
intro("hello","world")
# this will return an error that positional arguments are not allowed.
```

```python
def intro(**data):
    print("\nData type of argument:",type(data))
    print(data)
```

```python
intro(Firstname="Sita", Lastname="Sharma", Age=22, Phone=1234567890)
```

```python
# the input data is converted into a dict object, so can be parsed like one
def intro(**data):
    print("\nData type of argument:",type(data))
    for key, value in data.items():
        print(f"{key} is {value}")
```

```python
intro(Firstname="John", Lastname="Wood", Email="johnwood@nomail.com", Country="Wakanda", Age=25, Phone=9876543210)
```

```python
# if we are processing marks of students, this can be handy. we can specify marks along with the subject name. also, this will
# allow flexibility to handle varying number of subjects for each student.
```
