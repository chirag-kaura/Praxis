```python
# Exercise
# A student scored the following marks in her XII 100, 90, 96, 98, 85
# set this up as a list and parse using a loop and calculate the total marks

marks =[100, 90, 96, 98, 85]
total_marks =0
for m in marks:
    total_marks+=+m
print(total_marks)
```

```python
# calculate the standard deviation of marks
square_diff = 0
for m in marks:
    square_diff +=(m-avg_marks)**2
std_dev = (square_diff/len(marks))**0.5
print(std_dev)
```

```python
# Exercise
# print the frequency of each digit in a given number
# for e.g. if the number is 642130722 then, 0 occurs 1 time, 1 occurs 1 time, 2 occurs 3 times, ...
num = 642130722
freq ={str(i):0 for i in range(10)}

for ch in str(num):
    freq[ch]+=1

for digit,count in freq.items():
    if count>0:
        print(f"{digit} comes {count} times")
```

```python
#Calculate the monthly water bill using the following plan
#For first 50 units Rs. 0.50/unit
#For next 100 units Rs. 0.75/unit
#For next 100 units Rs. 1.20/unit
#For unit above 250 Rs. 1.50/unit
#An additional surcharge of 20% is added to the bill

# the customer used 513 units in the month
units =513
bill = 0
for u in range(1, units+1):
    if u <= 50:
        bill += 0.5
    elif u <= 150:
        bill += 0.75
    elif u <= 250:
        bill += 1.2
    else:
        bill += 1.5

print("Bill before surcharge is Rs.", bill)
# Apply the additional surcharge of 20%
bill *= 1.2
print("The monthly water bill is Rs.", bill)
```

```python
# Exercise
# A user requests password reset. Use has to prove their identity using OTP. The user is allowed a fixed number of tries.

# hint: while loop, fixed number of loops. within each loop, ask the user for OTP. if matches then break the loop and exit.
# if OTP does not match then use 'continue' and ask the user for OTP again.
correct_otp = "123456"
max_attempts = 3
attempts = 0 
while attempts < max_attempts:
    user_otp = input("Enter the OTP sent to your registered mobile number: ")
    if user_otp == correct_otp:
        print("OTP verified successfully. You can reset your password now.")
        break
    else:
        print("Incorrect OTP. Please try again.")
        attempts += 1
else:
    print("Maximum attempts reached. Please request a new OTP and try again.")
```
